# Scraping NFL Combine Stats & Inputting into DataFrame

## 2019 NFL Class

```python
url = 'https://www.pro-football-reference.com/play-index/nfl-combine-results.cgi?request=1&year_min=2019&year_max=2019&pos%5B%5D=WR&show=all&order_by=year_id'
res = requests.get(url)

res.status_code

soup = BeautifulSoup(res.content, 'lxml')
```

```python
combine_list = []

for i in soup.find('tbody').find_all('tr'):
    cells = i.find_all('td')
    combine_dict = {}
    combine_dict['name'] = cells[1].text
    combine_dict['age'] = cells[3].text
    combine_dict['school'] = cells[5].text
    combine_dict['height'] = cells[7].text
    combine_dict['weight'] = cells[8].text
    combine_dict['forty'] = cells[9].text
    combine_dict['vertical'] = cells[10].text
    combine_dict['bench'] = cells[11].text
    combine_dict['jump'] = cells[12].text
    combine_dict['cone'] = cells[13].text
    combine_dict['shuttle'] = cells[14].text
    combine_dict['drafted'] = cells[15].text
    combine_list.append(combine_dict)
```

### Correcting name inconsistencies between combine data and college data

```python
combine_list[44]['name']='JJ Arcega-Whiteside'
```

```python
combine_list[17]['name']='KeeSean Johnson'
```

```python
combine2019_df = pd.DataFrame(combine_list)
```

```python
combine2019_df.set_index('name',inplace=True)
```
